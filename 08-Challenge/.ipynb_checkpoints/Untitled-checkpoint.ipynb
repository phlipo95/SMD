```python
#print(iris['data'])         reinschauen in die dic "Seite" data
#print(iris['target'])       reinschauen in die dic "Seite" target
X, Y = iris.data[:, :2], iris.target 
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.Paired)
plt.show()
```
